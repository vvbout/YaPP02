## Предобработка данных

```python
# проверяем количество пропусков
data_null = data.isna().sum()
print(data_null)
```

```python
# анализируем пропуски в колонке *ceiling_height*
ceiling_height_null = data['ceiling_height'].isna().sum()
print("Количество пропущенных значений в колонке ceiling_height до обработки:", ceiling_height_null)
print("Процент пропущенных значений:", data['ceiling_height'].isna().mean())
print()

# создаем список идентификаторов удаленности от центра
cc_distance_id = ["2 км", "4 км", "6 км", "8 км", "10 км", "больше 10 км"]

# создаем функцию привязки удаллености и идентификатора
def cc_distance_id (cityCenters_nearest):
    if cityCenters_nearest <= 2000:
        return "2 км"
    if cityCenters_nearest <= 4000:
        return "4 км"
    if cityCenters_nearest <= 6000:
        return "6 км"
    if cityCenters_nearest <= 8000:
        return "8 км"
    if cityCenters_nearest <= 10000:
        return "10 км"
    else:
        return "больше 10 км"

# добавляем в таблицу колонку cc_distance_id
# проверяем кол-во уникальных значений
data["cc_distance_id"] = data["cityCenters_nearest"]. apply(cc_distance_id)
print("Количество уникальных значений по идентификатору расстояния от центра:\n", data["cc_distance_id"].value_counts())
print()

# вычисляем среднюю высоту потолка в "соседних" домах (растояние до центра)
mean_ch_cc_distance_id = data.groupby("cc_distance_id")["ceiling_height"].mean()

# заменяем пропущенные значения в колонке ceiling_height
#в зависимости от удаленности от центра с учетом удаленности от аэропорта
for cc_distance_id in mean_ch_cc_distance_id.index:
    data.loc[data['cc_distance_id'] == cc_distance_id,'ceiling_height'] = data.loc[data['cc_distance_id'] == cc_distance_id,'ceiling_height'].fillna(mean_ch_cc_distance_id[cc_distance_id])

# проверяем наличие пропущенных значений после замены
print("Количество пропущенных значений в колонке ceiling_height после обработки:", data['ceiling_height'].isna().sum())
```

```python
# анализируем пропуски в колонке *floors_total*
floors_total_null = data['floors_total'].isna().sum()

print("Количество пропущенных значений в колонке floors_total до обработки:", floors_total_null)
print("Процент пропущенных значений:", data['floors_total'].isna().mean())
print()

# удаляем строки с пропущенными данными
data = data.dropna(subset=['floors_total'])

# проверяем результат удаления
print("Количество пропущенных значений в колонке floors_total после обработки:", data['floors_total'].isna().sum())
print()
```

```python
# анализируем пропуски в колонке *living_area*
living_area_null = data['living_area'].isna().sum()

print("Количество пропущенных значений в колонке living_area до обработки:", living_area_null)
print("Процент пропущенных значений:", data['living_area'].isna().mean())

data['living_area'] = data['living_area'].fillna(0)
gb_liv_ar = data.groupby(['total_area','rooms'])
data['living_area'] = data['living_area'].fillna(gb_liv_ar['living_area'].transform(lambda group: group.mean() / (group.name[0])) * data['total_area'])

# проверяем результат замены
print("Количество пропущенных значений в колонке living_area после обработки:", data['living_area'].isna().sum())
```

```python
# анализируем пропуски в колонке *is_apartment*
is_apartment_null = data['is_apartment'].isna().sum()

print("Количество пропущенных значений в колонке is_apartment до обработки:", is_apartment_null)
print("Процент пропущенных значений:", data['is_apartment'].isna().mean())
print()

# проверяем уникальные значения
is_apartment_unique = data['is_apartment'].value_counts()
print("Количество уникальных значений до обработки:\n", is_apartment_unique)
print()

# производим замену, проверяем результат
data['is_apartment'].fillna(False, inplace=True)
print("Количество пропущенных значений после обработки:\n", data['is_apartment'].isna().sum())
print("Количество уникальных значений после обработки:\n", data['is_apartment'].value_counts())
```

```python
# анализируем пропуски в колонке *kitchen_area*
kitchen_area_null = data['kitchen_area'].isna().sum()

print("Количество пропущенных значений в колонке kitchen_area до обработки:", kitchen_area_null)
print("Процент пропущенных значений:", data['kitchen_area'].isna().mean())
print()

# проверяем корреляции
display(data.corr())
print()
print('Заметная сила корреляционной связи выявлена между kitchen_area и total area:{:.3f}'
      .format(data['kitchen_area'].corr(data['total_area'])))
print()
```

```python
# вычисляем индекс зависимости между total_area и kitchen_area
total_kitch_area_index = ((data['kitchen_area'].quantile(0.25) / data['total_area'].quantile(0.25))
                          + (data['kitchen_area'].quantile(0.50) / data['total_area'].quantile(0.50))
                          + (data['kitchen_area'].quantile(0.75) / data['total_area'].quantile(0.75)))/ 3

print('Индекс зависимости между total_area и kitchen_area:{:.3f}'.format(total_kitch_area_index))
```

```python
# производим замену пропущенных значений с учетом индекса
fillna_kit_area = data['total_area'] * total_kitch_area_index
data['kitchen_area'] = data['kitchen_area'].fillna(fillna_kit_area)
print("Количество пропущенных значений после обработки:", data['kitchen_area'].isna().sum())
print("Статистика колонки kitchen_area после обработки:\n",data['kitchen_area'].describe())
```

```python
# составляем рейтинг населенных пунктов,
# делаем срез
# рассчитываем среднюю цену по каждому пунтку
data_top_ten = data_filtered_ta_lp_r_ch['locality_name'].value_counts().head(10)
data_top_ten_flt = data_filtered_ta_lp_r_ch.query('locality_name in @data_top_ten.index')
display(data_top_ten_flt.pivot_table(index='locality_name', values='price_pm').round().sort_values(by='price_pm', ascending=False))
```

На первом месте по стоимости (**самая высокая**) закономерно оказался г. Санкт-Петербург. **Самая низкая** стоимость выявлена в г. Выборг.

#### Выделение центральной зоны в Санкт-Петербурге

```python
# делаем срез по Санкт-Петербургу,
data_spb_flt = data_filtered_ta_lp_r_ch.query('(locality_name == "санкт-петербург") & (cityCenters_nearest !=888)')
data_spb_flt = data_spb_flt.reset_index(drop=True)
display(data_spb_flt.head())
```

```python
# добавляем колонку с расстоянием до центра, переведенным в км
data_spb_flt['city_center_km'] = data_spb_flt['cityCenters_nearest'] / 1000
data_spb_flt['city_center_km'] = data_spb_flt['city_center_km'].astype('int64')

# проверяем минимальное значение
print("Минимальная удаленность от центра:", data_spb_flt['city_center_km'].min())
display(data_spb_flt.query('city_center_km <1'))
```

```python
# заменяем значение 0 на 1
data_spb_flt['city_center_km']=data_spb_flt['city_center_km'].replace(0, 1)

# сводим данные
data_spb_flt_pivot_km = data_spb_flt.pivot_table(index='city_center_km', values='price_pm')
display(data_spb_flt_pivot_km.head(10))
```

```python
# строим столбчатую диаграмму
data_spb_flt_pivot_km.plot(kind='bar', y= 'price_pm', ylim=(60000, 140000), figsize=(10, 5), grid=True, title='ВЛИЯНИЕ РАССТОЯНИЯ ОТ ЦЕНТРА СПБ НА ЦЕНУ_М2')
plt.show
```

Санкт - Петербургу с помощью метода *.query()*, группировка данных и расчет значения средней стоимости м2 в зависимости от удаленности от центра города с помощью метода *.pivot_table()* позволили создать график с помощью метода *.plot(), на котором видно, что стоимость меняется на 5м км.
Принимаем радиус 5км от центра как **границу центральной зоны**, что совпадает с границей *Центрального района* г. Санкт - Петербурга (3 км) + 2 км зона застройки элитного жилья.
